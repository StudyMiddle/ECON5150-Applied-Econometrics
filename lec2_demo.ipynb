Example 1


We use the example in the lecture notes

$$
z_{n}=\begin{cases}
-\sqrt{n} & \text{with probability }1/n\\
0 & \text{with probability }1-2/n\\
\sqrt{n} & \text{with probability }\text{1/n}.
\end{cases}
$$

This example violates uniform integrability, and also violates Lindeberg condition.
CLT fails.
The standardized statistic does not converge to N(0,1) in distribution

```r
n = 10
Rep = 1000
Tot = n * Rep

x <- sqrt(n) * 
    (rbinom(Tot, 1, 1/n * 2)) *
    (2 * rbinom(Tot, 1, 0.5)  - 1 )

X <- matrix(x, Rep)

Sn = apply(X, 1, sum)
agg_sd <- sqrt( n * 2 )

Y = Sn/agg_sd

plot(density(Y), xlab = "")
```

Example 2

check the distribution of $\max_{i \leq n} |x_i|$, where $x_i \sim N(0,1)$



```r

TT = 1000

max_n <- function(n) {
  X = matrix( abs( rnorm(n*TT) ), TT)
  Y = apply(X, 1, max)
  return(Y)
}

plot( density(max_n(10)), ylim = c(0,1.6), xlab = "")
lines(density(max_n(100)), col="red")
lines(density(max_n(1000)), col="blue")
```
